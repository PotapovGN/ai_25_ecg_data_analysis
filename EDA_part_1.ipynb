```python

# Пути к данным
BASE = "/content/drive/MyDrive/EKG/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3"
DB_CSV = os.path.join(BASE, "ptbxl_database.csv")
SCP_CSV = os.path.join(BASE, "scp_statements.csv")
RECORDS_100 = os.path.join(BASE, "records100")
RECORDS_500 = os.path.join(BASE, "records500")

# Директории для чекпоинтов
CHECKPOINT_DIR = "/content/drive/MyDrive/EKG/checkpoints"
FILE_ORDER_PATH = os.path.join(CHECKPOINT_DIR, "file_order.pkl")
PROGRESS_PATH = os.path.join(CHECKPOINT_DIR, "progress.pkl")
os.makedirs(CHECKPOINT_DIR, exist_ok=True)

# Параметры отработки
START_FROM = 0  # Измените этот параметр для возобновления с определенной записи
CHUNK_SIZE = 500  # Сохранять каждые 500 записей
N_SAMPLES = None  # Сколько записей обработать (None = все)

# Загруска CSV
df_db = pd.read_csv(DB_CSV, low_memory=False)
df_scp = pd.read_csv(SCP_CSV, index_col=0, low_memory=False)

# Начало блока кода, разработанного с использованием ChatGPT (OpenAI)
# Преобразование scp_codes из строки в словарь
def parse_scp_codes(scp_str):
    """Парсинг scp_codes из строки в словарь"""
    try:
        if pd.isna(scp_str):
            return {}
        scp_str = scp_str.replace("'", "\"")
        return json.loads(scp_str)
    except:
        return {}

df_db['scp_codes_parsed'] = df_db['scp_codes'].apply(parse_scp_codes)


# ФУНКЦИИ ИЗВЛЕЧЕНИЯ ПРИЗНАКОВ (ВАШ ИСХОДНЫЙ КОД БЕЗ ИЗМЕНЕНИЙ)
def compute_hrv_features(signal, fs):
    """Улучшенные HRV признаки с обработкой коротких записей"""
    try:
        _, rpeaks = nk.ecg_peaks(signal, sampling_rate=fs)

        if len(rpeaks['ECG_R_Peaks']) < 4:
            return {k: np.nan for k in ['meanRR','sdNN','RMSSD','pNN50','LF','HF','LF_HF']}

        hrv_time = nk.hrv_time(rpeaks, sampling_rate=fs, show=False)

        if len(rpeaks['ECG_R_Peaks']) >= 10:
            hrv_freq = nk.hrv_frequency(rpeaks, sampling_rate=fs, show=False)
        else:
            hrv_freq = pd.DataFrame()

        def safe_val(df, key):
            return df[key].values[0] if key in df.columns and len(df) > 0 else np.nan

        return {
            'meanRR': safe_val(hrv_time, 'HRV_MeanNN'),
            'sdNN': safe_val(hrv_time, 'HRV_SDNN'),
            'RMSSD': safe_val(hrv_time, 'HRV_RMSSD'),
            'pNN50': safe_val(hrv_time, 'HRV_pNN50'),
            'LF': safe_val(hrv_freq, 'HRV_LF'),
            'HF': safe_val(hrv_freq, 'HRV_HF'),
            'LF_HF': safe_val(hrv_freq, 'HRV_LFHF')
        }
    except Exception as e:
        return {k: np.nan for k in ['meanRR','sdNN','RMSSD','pNN50','LF','HF','LF_HF']}

def compute_robust_morphological_features(signal, fs):
    """НАДЕЖНОЕ извлечение морфологических признаков с альтернативными методами"""
    try:
        features = {}
        cleaned = nk.ecg_clean(signal, sampling_rate=fs)
        _, rpeaks = nk.ecg_peaks(cleaned, sampling_rate=fs)

        if len(rpeaks['ECG_R_Peaks']) < 2:
            return create_nan_morphology_features()

        r_peaks = rpeaks['ECG_R_Peaks']
        features['R_amp'] = np.nanmean(cleaned[r_peaks])

        delineate_info = try_multiple_delineate_methods(cleaned, rpeaks, fs)

        if delineate_info:
            features.update(extract_detailed_morphology(cleaned, rpeaks, delineate_info, fs))
        else:
            features.update(extract_basic_morphology(cleaned, r_peaks, fs))

        features.update(simple_st_analysis(cleaned, r_peaks, fs))
        return features

    except Exception as e:
        return create_nan_morphology_features()

def try_multiple_delineate_methods(signal, rpeaks, fs):
    """Пробует разные методы delineate"""
    methods = ["dwt", "cwt", "peak"]
    for method in methods:
        try:
            _, delineate_info = nk.ecg_delineate(signal, rpeaks, sampling_rate=fs, method=method, show=False)
            if (len(delineate_info.get('ECG_P_Peaks', [])) > 0 or
                len(delineate_info.get('ECG_T_Peaks', [])) > 0):
                return delineate_info
        except:
            continue
    return None

def extract_detailed_morphology(signal, rpeaks, delineate_info, fs):
    """Извлечение детальных морфологических признаков когда delineate работает"""
    features = {}
    wave_types = ['P', 'Q', 'R', 'S', 'T']
    for wave in wave_types:
        peaks = delineate_info.get(f'ECG_{wave}_Peaks', [])
        if len(peaks) > 0:
            features[f'{wave}_amp'] = np.nanmean(signal[peaks])
        else:
            features[f'{wave}_amp'] = np.nan

    try:
        intervals = nk.ecg_interval(signal, rpeaks, sampling_rate=fs)
        if 'ECG_Q_Durations' in intervals and 'ECG_S_Durations' in intervals:
            q_durations = intervals['ECG_Q_Durations']
            s_durations = intervals['ECG_S_Durations']
            if len(q_durations) > 0 and len(s_durations) > 0:
                features['QRS_dur'] = np.nanmean(q_durations + s_durations)

        if 'ECG_T_Durations' in intervals:
            qt_intervals = intervals['ECG_T_Durations']
            if len(qt_intervals) > 0:
                features['QT_int'] = np.nanmean(qt_intervals)

        if 'ECG_P_Durations' in intervals:
            pr_intervals = intervals['ECG_P_Durations']
            if len(pr_intervals) > 0:
                features['PR_int'] = np.nanmean(pr_intervals)
    except:
        pass

    if not np.isnan(features.get('P_amp', np.nan)) and not np.isnan(features.get('T_amp', np.nan)):
        features['PT_ratio'] = features['P_amp'] / (features['T_amp'] + 1e-6)

    if not np.isnan(features.get('Q_amp', np.nan)) and not np.isnan(features.get('R_amp', np.nan)):
        features['QR_ratio'] = abs(features['Q_amp']) / (features['R_amp'] + 1e-6)

    return features

def extract_basic_morphology(signal, r_peaks, fs):
    """Упрощенный анализ морфологии когда delineate не работает"""
    features = {
        'P_amp': np.nan, 'Q_amp': np.nan, 'S_amp': np.nan, 'T_amp': np.nan,
        'QRS_dur': np.nan, 'QT_int': np.nan, 'PR_int': np.nan,
        'PT_ratio': np.nan, 'QR_ratio': np.nan
    }

    try:
        qrs_durations = []
        for r_peak in r_peaks:
            if r_peak > 50 and r_peak < len(signal) - 50:
                segment = signal[r_peak-25:r_peak+25]
                diff_segment = np.gradient(segment)
                threshold = 0.1 * np.max(np.abs(diff_segment))
                qrs_points = np.where(np.abs(diff_segment) > threshold)[0]
                if len(qrs_points) > 0:
                    qrs_dur = (qrs_points[-1] - qrs_points[0]) / fs * 1000
                    qrs_durations.append(qrs_dur)

        if qrs_durations:
            features['QRS_dur'] = np.nanmean(qrs_durations)
    except:
        pass

    return features

def simple_st_analysis(signal, r_peaks, fs):
    """Упрощенный анализ ST сегмента"""
    try:
        st_segments = []
        st_slopes = []

        for r_peak in r_peaks:
            if r_peak < len(signal) - int(0.2 * fs):
                j_point = min(r_peak + int(0.04 * fs), len(signal) - 1)
                st60_point = min(j_point + int(0.06 * fs), len(signal) - 1)

                j_amp = signal[j_point]
                st60_amp = signal[st60_point]

                st_segment = st60_amp - j_amp
                st_segments.append(st_segment)

                if st60_point > j_point:
                    st_slope = (st60_amp - j_amp) / 0.06
                    st_slopes.append(st_slope)

        return {
            'ST_segment': np.nanmean(st_segments) if st_segments else np.nan,
            'ST_slope': np.nanmean(st_slopes) if st_slopes else np.nan,
            'J_point_amp': np.nanmean([signal[min(r + int(0.04 * fs), len(signal)-1)] for r in r_peaks]) if r_peaks else np.nan,
            'ST60_amp': np.nanmean([signal[min(r + int(0.1 * fs), len(signal)-1)] for r in r_peaks]) if r_peaks else np.nan
        }
    except:
        return {
            'ST_segment': np.nan, 'ST_slope': np.nan,
            'J_point_amp': np.nan, 'ST60_amp': np.nan
        }

def create_nan_morphology_features():
    """Создает все морфологические признаки с NaN"""
    base_features = ['P_amp', 'Q_amp', 'R_amp', 'S_amp', 'T_amp',
                    'QRS_dur', 'QT_int', 'PR_int', 'ST_segment',
                    'ST_slope', 'J_point_amp', 'ST60_amp',
                    'PT_ratio', 'QR_ratio']
    return {feature: np.nan for feature in base_features}

def compute_frequency_features(signal, fs):
    """Частотные признаки на основе PSD и STFT"""
    try:
        f, Pxx = welch(signal, fs=fs, nperseg=min(fs*2, len(signal)//2))
        f_mean = np.sum(f * Pxx) / np.sum(Pxx)
        f_std = np.sqrt(np.sum(((f - f_mean)**2) * Pxx) / np.sum(Pxx))

        _, _, Zxx = stft(signal, fs=fs, nperseg=min(256, len(signal)//4))
        stft_energy = np.mean(np.abs(Zxx)**2)

        return {
            'PSD_mean_freq': f_mean,
            'PSD_std_freq': f_std,
            'STFT_energy': stft_energy,
            'PSD_power_mean': np.mean(Pxx),
            'PSD_power_max': np.max(Pxx)
        }
    except Exception:
        return {k: np.nan for k in ['PSD_mean_freq','PSD_std_freq','STFT_energy','PSD_power_mean','PSD_power_max']}

def compute_wavelet_features(signal):
    """Вейвлет признаки (энергии уровней)"""
    try:
        coeffs = pywt.wavedec(signal, 'db4', level=4)
        energies = {f'wavelet_E{i+1}': np.sum(np.square(c)) for i, c in enumerate(coeffs)}
        entropy = antropy.spectral_entropy(signal, sf=500, method='welch')
        return {**energies, 'wavelet_entropy': entropy}
    except Exception:
        return {'wavelet_E1': np.nan, 'wavelet_E2': np.nan, 'wavelet_E3': np.nan, 'wavelet_E4': np.nan, 'wavelet_entropy': np.nan}

def compute_signal_quality(signal):
    """Метрики качества сигнала"""
    try:
        signal_det = detrend(signal)
        noise_power = np.mean((signal - signal_det)**2)
        if noise_power > 0:
            snr = 10 * np.log10(np.mean(signal**2) / noise_power)
        else:
            snr = np.nan
        prop_censored = np.mean(np.abs(signal) > 5*np.std(signal))
        return {'SNR_dB': snr, 'prop_censored': prop_censored}
    except Exception:
        return {'SNR_dB': np.nan, 'prop_censored': np.nan}

def compute_myocardial_features(signal, fs):
    """Признаки для диагностики инфаркта миокарда"""
    try:
        cleaned = nk.ecg_clean(signal, sampling_rate=fs)
        _, rpeaks = nk.ecg_peaks(cleaned, sampling_rate=fs)

        features = {
            'Q_wave_amp': np.nan,
            'Q_wave_dur': np.nan,
            'pathological_Q': False
        }

        q_amplitudes = []
        for r_peak in rpeaks['ECG_R_Peaks']:
            if r_peak > 20:
                search_start = max(0, r_peak - int(0.08 * fs))
                search_end = r_peak
                segment = cleaned[search_start:search_end]
                if len(segment) > 0:
                    min_idx = np.argmin(segment)
                    q_amplitude = segment[min_idx]
                    if q_amplitude < -0.05:
                        q_amplitudes.append(abs(q_amplitude))

        if q_amplitudes:
            features.update({
                'Q_wave_amp': np.nanmean(q_amplitudes),
                'Q_wave_dur': np.nan,
                'pathological_Q': np.nanmean(q_amplitudes) > 0.1
            })

        return features

    except Exception:
        return {
            'Q_wave_amp': np.nan,
            'Q_wave_dur': np.nan,
            'pathological_Q': False
        }

def extract_features_from_ecg_record(ecg_record, fs):
    """Объединение всех улучшенных признаков"""
    features = {}

    for lead_name, signal in ecg_record.items():
        lead_features = {}

        lead_features.update(compute_hrv_features(signal, fs))
        lead_features.update(compute_robust_morphological_features(signal, fs))
        lead_features.update(compute_frequency_features(signal, fs))
        lead_features.update(compute_wavelet_features(signal))
        lead_features.update(compute_signal_quality(signal))

        if lead_name in ['V1', 'V2', 'V3', 'V4', 'V5', 'V6']:
            lead_features.update(compute_myocardial_features(signal, fs))

        features.update({f"{lead_name}_{k}": v for k, v in lead_features.items()})

    return features

def add_scp_descriptions(scp_codes_dict, df_scp):
    """Добавляет расшифровку SCP-кодов"""
    descriptions = {}

    for scp_code, confidence in scp_codes_dict.items():
        if scp_code in df_scp.index:
            scp_info = df_scp.loc[scp_code]
            prefix = f"scp_{scp_code}"

            descriptions.update({
                f"{prefix}_description": scp_info.get('description', ''),
                f"{prefix}_diagnostic": scp_info.get('diagnostic', np.nan),
                f"{prefix}_form": scp_info.get('form', np.nan),
                f"{prefix}_rhythm": scp_info.get('rhythm', np.nan),
                f"{prefix}_diagnostic_class": scp_info.get('diagnostic_class', ''),
                f"{prefix}_diagnostic_subclass": scp_info.get('diagnostic_subclass', ''),
                f"{prefix}_confidence": confidence
            })

    return descriptions

# СИСТЕМА ЧЕКПОИНТОВ И ВОЗОБНОВЛЕНИЯ (ОБНОВЛЕННАЯ)
def save_file_order():
    """Сохраняет порядок обработки файлов с индексами"""
    file_order = {
        'ecg_ids': df_db['ecg_id'].tolist(),
        'indices': df_db.index.tolist(),
        'timestamp': pd.Timestamp.now()
    }
    with open(FILE_ORDER_PATH, 'wb') as f:
        pickle.dump(file_order, f)
    print(f"✅ Порядок обработки сохранен: {len(file_order['ecg_ids'])} записей")

def load_file_order():
    """Загружает порядок обработки файлов"""
    if os.path.exists(FILE_ORDER_PATH):
        try:
            # Проверяем, что файл не пустой
            if os.path.getsize(FILE_ORDER_PATH) > 0:
                with open(FILE_ORDER_PATH, 'rb') as f:
                    return pickle.load(f)
            else:
                print("⚠️ Файл порядка обработки пустой, создаем новый")
                return None
        except (EOFError, pickle.UnpicklingError) as e:
            print(f"⚠️ Ошибка загрузки файла порядка: {e}, создаем новый")
            return None
    return None
def save_progress(current_index, processed_ecg_ids, errors):
    """Сохраняет текущий прогресс с обработанными ECG_ID"""
    progress = {
        'current_index': current_index,
        'processed_ecg_ids': list(processed_ecg_ids),  # Сохраняем какие файлы уже обработаны
        'errors': errors,
        'timestamp': pd.Timestamp.now()
    }
    with open(PROGRESS_PATH, 'wb') as f:
        pickle.dump(progress, f)

def load_progress():
    """Загружает последний прогресс"""
    if os.path.exists(PROGRESS_PATH):
        with open(PROGRESS_PATH, 'rb') as f:
            return pickle.load(f)
    return None

def save_chunk(chunk_id, features_chunk, errors_chunk):
    """Сохраняет чанк данных"""
    chunk_filename = f"features_chunk_{chunk_id:04d}.csv"
    chunk_path = os.path.join(CHECKPOINT_DIR, chunk_filename)

    df_chunk = pd.DataFrame(features_chunk)
    df_chunk.to_csv(chunk_path, index=False)

    # Сохраняем ошибки чанка
    if errors_chunk:
        errors_filename = f"errors_chunk_{chunk_id:04d}.txt"
        errors_path = os.path.join(CHECKPOINT_DIR, errors_filename)
        with open(errors_path, 'w') as f:
            for error in errors_chunk:
                f.write(error + '\n')

    print(f"✅ Чанк {chunk_id} сохранен: {chunk_path}")

def merge_all_chunks():
    """Объединяет все чанки в один файл"""
    chunk_files = [f for f in os.listdir(CHECKPOINT_DIR) if f.startswith('features_chunk_') and f.endswith('.csv')]
    chunk_files.sort()

    all_dfs = []
    for chunk_file in chunk_files:
        chunk_path = os.path.join(CHECKPOINT_DIR, chunk_file)
        df_chunk = pd.read_csv(chunk_path)
        all_dfs.append(df_chunk)
        print(f"Загружен чанк: {chunk_file} ({len(df_chunk)} записей)")

    if all_dfs:
        df_final = pd.concat(all_dfs, ignore_index=True)
        final_path = "/content/drive/MyDrive/EKG/ecg_features_final.csv"
        df_final.to_csv(final_path, index=False)
        print(f"✅ Финальный файл создан: {final_path} ({len(df_final)} записей)")
        return df_final
    else:
        print("❌ Нет чанков для объединения")
        return None

# ОСНОВНАЯ ОБРАБОТКА
# Инициализация системы чекпоинтов
file_order_data = load_file_order()
if not file_order_data:
    save_file_order()
    file_order_data = load_file_order()

print(f"Всего записей для обработки: {len(df_db)}")

# Загрузка прогресса или инициализация
progress = load_progress()
if progress and START_FROM == 0:
    print("🔄 Найден предыдущий прогресс, возобновляем...")
    current_index = progress['current_index']
    processed_ecg_ids = set(progress['processed_ecg_ids'])
    errors = progress['errors']
    print(f"Прогресс: {current_index} записей обработано, {len(processed_ecg_ids)} записей в processed_ecg_ids")

    # ОТЛАДОЧНАЯ ИНФОРМАЦИЯ
    print(f"Текущий индекс: {current_index}, Всего записей в БД: {len(df_db)}")
    if current_index >= len(df_db):
        print("⚠️ ВНИМАНИЕ: current_index превышает количество записей в БД!")
        current_index = min(current_index, len(df_db) - 1)
else:
    print("🚀 Начинаем новую обработку...")
    current_index = START_FROM
    processed_ecg_ids = set()
    errors = []
    if START_FROM > 0:
        print(f"Начинаем с записи {START_FROM}")

# Основной цикл обработки
chunk_id = current_index // CHUNK_SIZE
processed_in_chunk = 0
all_features = []

print(f"Начало обработки ЭКГ записей...")

for i in tqdm(range(current_index, len(df_db))):
    if N_SAMPLES is not None and len(processed_ecg_ids) >= N_SAMPLES:
        break

    row = df_db.iloc[i]
    ecg_id = row['ecg_id']

    # Пропускаем уже обработанные файлы
    if ecg_id in processed_ecg_ids:
        print(f"Пропускаем уже обработанный файл: {ecg_id}")
        continue

    # Поиск файла
    rec_path_500 = os.path.join(BASE, row['filename_hr'])
    rec_path_100 = os.path.join(BASE, row['filename_lr'])

    if os.path.exists(rec_path_500 + ".dat"):
        rec_path = rec_path_500
    elif os.path.exists(rec_path_100 + ".dat"):
        rec_path = rec_path_100
    else:
        errors.append(f"Файл не найден: {rec_path_500} и {rec_path_100}")
        continue

    try:
        # Чтение ЭКГ записи
        record = wfdb.rdrecord(rec_path)
        ecg_record = {lead: record.p_signal[:, j] for j, lead in enumerate(record.sig_name)}

        # Извлечение признаков сигнала
        feats = extract_features_from_ecg_record(ecg_record, fs=record.fs)

        # Добавление метаданных пациента
        patient_meta = {
            'ecg_id': row['ecg_id'],
            'patient_id': row['patient_id'],
            'age': row['age'],
            'sex': row['sex'],
            'height': row['height'],
            'weight': row['weight'],
            'nurse': row['nurse'],
            'site': row['site'],
            'device': row['device'],
            'recording_date': row['recording_date'],
            'report': row['report'],
            'heart_axis': row['heart_axis'],
            'infarction_stadium1': row['infarction_stadium1'],
            'infarction_stadium2': row['infarction_stadium2'],
            'validated_by': row['validated_by'],
            'second_opinion': row['second_opinion'],
            'initial_autogenerated_report': row['initial_autogenerated_report'],
            'validated_by_human': row['validated_by_human'],
            'baseline_drift': row['baseline_drift'],
            'static_noise': row['static_noise'],
            'burst_noise': row['burst_noise'],
            'electrodes_problems': row['electrodes_problems'],
            'extra_beats': row['extra_beats'],
            'pacemaker': row['pacemaker'],
            'strat_fold': row['strat_fold'],
            'filename_lr': row['filename_lr'],
            'filename_hr': row['filename_hr'],
            'scp_codes_raw': row['scp_codes']  # оригинальная строка
        }

        # Добавляем параметры записи
        recording_meta = {
            'signal_fs': record.fs,
            'signal_length': record.sig_len,
            'n_leads': record.n_sig,
            'lead_names': ','.join(record.sig_name)
        }

        # Добавляем расшифровку SCP-кодов
        scp_descriptions = add_scp_descriptions(row['scp_codes_parsed'], df_scp)

        # Объединяем все признаки
        feats.update(patient_meta)
        feats.update(recording_meta)
        feats.update(scp_descriptions)

        all_features.append(feats)
        processed_ecg_ids.add(ecg_id)  # Отмечаем как обработанный
        processed_in_chunk += 1

        # Сохранение чанка каждые CHUNK_SIZE записей
        if processed_in_chunk >= CHUNK_SIZE:
            chunk_id = i // CHUNK_SIZE
            save_chunk(chunk_id, all_features, errors)
            all_features = []
            processed_in_chunk = 0
            # Сохраняем прогресс с обновленным списком обработанных файлов
            save_progress(i + 1, processed_ecg_ids, errors)
            print(f"✅ Прогресс сохранен: {i + 1} записей")

    except Exception as e:
        error_msg = f"Ошибка при обработке {rec_path}: {str(e)}"
        errors.append(error_msg)
        continue

# Сохранение последнего чанка
if all_features:
    chunk_id = len(df_db) // CHUNK_SIZE
    save_chunk(chunk_id, all_features, errors)

# Удаление файла прогресса после успешного завершения
if os.path.exists(PROGRESS_PATH):
    os.remove(PROGRESS_PATH)
    print("✅ Файл прогресса удален после успешного завершения")

# Объединение всех чанков
print("🔄 Объединение всех чанков...")
df_final = merge_all_chunks()

print("\n🎯 ОБРАБОТКА ЗАВЕРШЕНА!")
if df_final is not None:
    print(f"Общее количество обработанных записей: {len(df_final)}")
    print(f"Финальный файл: /content/drive/MyDrive/EKG/ecg_features_final.csv")

    # Выводим информацию о структуре данных
    print("\n📊 СТРУКТУРА СОХРАНЕННЫХ ДАННЫХ:")
    print(f"Всего признаков: {len(df_final.columns)}")
    print(f"Всего записей: {len(df_final)}")

    # Проверяем наличие ключевых морфологических признаков
    morphology_cols = [col for col in df_final.columns if any(wave in col for wave in ['_P_amp', '_R_amp', '_T_amp', '_QRS_dur', '_QT_int'])]
    print(f"Морфологических признаков: {len(morphology_cols)}")

    # Проверяем заполненность ключевых признаков
    if len(df_final) > 0:
        sample_morphology = df_final.iloc[0][morphology_cols[:5]]  # Первые 5 морфологических признаков
        print("\nПример морфологических признаков (первая запись):")
        for col, val in sample_morphology.items():
            print(f"  {col}: {val}")
# Конец блока кода, разработанного с использованием ChatGPT (OpenAI)
```
