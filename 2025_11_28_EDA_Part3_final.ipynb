```python
def create_col_list_from_search_string(df, *search_strings):
    """
    Создает список признаков по нескольким строкам поиска в названии
    """
    found_columns = []

    for search_string in search_strings:
        # Находим все признаки содержащие текущую строку поиска в названии
        columns_for_string = [col for col in df.columns if search_string in col]
        found_columns.extend(columns_for_string)

    # Убираем дубликаты на случай пересечения поисковых строк
    found_columns = list(set(found_columns))
    print(f"Найдено признаков - {len(found_columns)}")
    return found_columns
```

Интерпретация: Чем выше - лучше работа парасимпатической нервной системы

```python
from scipy.stats import pointbiserialr
from sklearn.feature_selection import mutual_info_classif, f_classif

#функция написана с помощью Deepseek
def analyze_feature_target_correlations(feature_columns, df=df, target_columns=target_columns):
    """
    Анализирует взаимосвязь между количественными признаками и бинарными целевыми переменными

    Parameters:
    -----------
    df : pandas.DataFrame
        Датафрейм с данными
    feature_columns : list
        Список названий количественных признаков
    target_columns : list
        Список названий бинарных целевых переменных

    Returns:
    --------
    pandas.DataFrame
        Датафрейм с результатами анализа
    """

    results = []

    for feature in feature_columns:
        for target in target_columns:
            # Проверяем, что данные существуют
            if feature not in df.columns or target not in df.columns:
                continue

            # Убираем пропущенные значения
            mask = ~(df[feature].isna() | df[target].isna())
            x = df.loc[mask, feature]
            y = df.loc[mask, target]

            # Point-biserial correlation
            try:
                pb_corr, pb_pvalue = pointbiserialr(x, y)
            except:
                pb_corr, pb_pvalue = np.nan, np.nan

            # Mutual Information
            try:
                # Reshape для sklearn
                x_reshaped = x.values.reshape(-1, 1)
                mi_score = mutual_info_classif(x_reshaped, y, random_state=42)[0]
            except:
                mi_score = np.nan

            # ANOVA F-value
            try:
                f_stat, f_pvalue = f_classif(x_reshaped, y)
                f_value = f_stat[0]
            except:
                f_value, f_pvalue = np.nan, np.nan

            # Сохраняем результаты
            results.append({
                'feature': feature,
                'target': target,
                'point_biserial_corr': pb_corr,
                'point_biserial_pvalue': pb_pvalue,
                'mutual_info': mi_score,
                'f_value': f_value,
                'f_pvalue': f_pvalue
            })

    # Создаем итоговый датафрейм
    results_df = pd.DataFrame(results)

    # Выводим результаты
    print("Анализ взаимосвязи признаков с целевыми переменными:")
    print("=" * 80)
    print(f"Количество признаков: {len(feature_columns)}")
    print(f"Количество целевых переменных: {len(target_columns)}")
    print(f"Всего комбинаций: {len(results_df)}")
    print("\n" + "=" * 80)

    # Форматируем вывод для лучшей читаемости
    pd.set_option('display.max_rows', None)
    pd.set_option('display.width', None)

    # Сортируем по mutual information (по убыванию)
    results_df_sorted = results_df.sort_values(['target', 'mutual_info'], ascending=[True, False])

    print(results_df_sorted.round(4))
```

```python
#функция написана с помощью Deepseek

def print_correlation_matrices(feature_columns, df=df):
    """
    Создает и выводит матрицы попарных корреляций (Пирсон и Спирман)

    Parameters:
    -----------
    df : pandas.DataFrame
        Датафрейм с данными
    feature_columns : list
        Список названий признаков для анализа корреляций
    """

    # Выбираем только нужные колонки
    data_subset = df[feature_columns]

    # Вычисляем матрицу корреляций Пирсона
    pearson_corr = data_subset.corr(method='pearson')

    # Вычисляем матрицу корреляций Спирмана
    spearman_corr = data_subset.corr(method='spearman')

    # Выводим матрицу Пирсона
    print("=" * 80)
    print("МАТРИЦА КОРРЕЛЯЦИЙ ПИРСОНА")
    print("=" * 80)
    print(pearson_corr.round(4))

    print("\n" + "=" * 80)
    print("МАТРИЦА КОРРЕЛЯЦИЙ СПИРМАНА")
    print("=" * 80)
    print(spearman_corr.round(4))

    # Дополнительная информация о высоких корреляциях
    print("\n" + "=" * 80)
    print("ИНФОРМАЦИЯ О ВЫСОКИХ КОРРЕЛЯЦИЯХ (|r| > 0.8)")
    print("=" * 80)

    # Анализ высоких корреляций для Пирсона
    high_corr_pearson = []
    high_corr_spearman = []

    # Собираем пары с высокой корреляцией (исключая диагональ)
    for i in range(len(pearson_corr.columns)):
        for j in range(i + 1, len(pearson_corr.columns)):
            pearson_val = pearson_corr.iloc[i, j]
            spearman_val = spearman_corr.iloc[i, j]

            if abs(pearson_val) > 0.7:
                high_corr_pearson.append({
                    'feature1': pearson_corr.columns[i],
                    'feature2': pearson_corr.columns[j],
                    'correlation': pearson_val
                })

            if abs(spearman_val) > 0.7:
                high_corr_spearman.append({
                    'feature1': spearman_corr.columns[i],
                    'feature2': spearman_corr.columns[j],
                    'correlation': spearman_val
                })

    # Сортируем по убыванию корреляции
    high_corr_pearson.sort(key=lambda x: abs(x['correlation']), reverse=True)
    high_corr_spearman.sort(key=lambda x: abs(x['correlation']), reverse=True)

    print(f"\nВысокие корреляции Пирсона (> 0.7): {len(high_corr_pearson)} пар")
    for pair in high_corr_pearson[:10]:  # Показываем топ-10
        print(f"  {pair['feature1']} - {pair['feature2']}: {pair['correlation']:.4f}")

    print(f"\nВысокие корреляции Спирмана (> 0.7): {len(high_corr_spearman)} пар")
    for pair in high_corr_spearman[:10]:  # Показываем топ-10
        print(f"  {pair['feature1']} - {pair['feature2']}: {pair['correlation']:.4f}")

# Пример использования:
# print_correlation_matrices(df, ['I_sdNN', 'II_sdNN', 'I_RMSSD', 'II_RMSSD'])
```

```python
#функция написана с помощью Deepseek
def filter_features(features_to_keep, checked_features=analyze_list, numeric_columns=numeric_columns):
    """
    Оставляет только выбранные признаки из проверенного списка, удаляя остальные

    Parameters:
    -----------
    numeric_columns : list
        Исходный список всех числовых признаков
    checked_features : list
        Список признаков, которые были проверены/анализированы
    features_to_keep : list
        Список признаков, которые нужно оставить

    Returns:
    --------
    list
        Новый список признаков после фильтрации
    """

    # Создаем копию исходного списка
    filtered_columns = numeric_columns.copy()

    # Удаляем все проверенные признаки, кроме тех, что нужно оставить
    for feature in checked_features:
        if feature not in features_to_keep:
            if feature in filtered_columns:
                filtered_columns.remove(feature)

    return filtered_columns

# Пример использования:
# numeric_columns = ['I_sdNN', 'II_RMSSD', 'V1_RMSSD', 'AVR_RMSSD', 'V3_sdNN', 'age', 'height']
# checked_features = ['I_sdNN', 'II_RMSSD', 'V1_RMSSD', 'AVR_RMSSD', 'V3_sdNN']
# features_to_keep = ['V1_RMSSD', 'AVR_RMSSD']
#
# new_columns = filter_features(numeric_columns, checked_features, features_to_keep)
# print(new_columns)  # ['V1_RMSSD', 'AVR_RMSSD', 'age', 'height']
```

Проверим можно ли вернуть какие то признаки

```python
from scipy.stats import pearsonr, spearmanr

def check_returnable_features():
    """
    Проверяет признаки из init_numeric_columns на возможность возвращения
    и выводит DataFrame с признаками, которые можно вернуть
    """
    # Создаем список для хранения результатов
    returnable_features = []

    # Проверка признаков из init_numeric_columns на возможность возвращения
    for feature in init_numeric_columns:
        if feature not in numeric_columns:
            max_correlation = 0
            can_return = True

            # Проверяем корреляции с текущими признаками в numeric_columns
            for existing_feature in numeric_columns:
                pearson_corr = abs(pearsonr(df[feature], df[existing_feature])[0])
                spearman_corr = abs(spearmanr(df[feature], df[existing_feature])[0])
                current_max_corr = max(pearson_corr, spearman_corr)

                # Обновляем максимальную корреляцию
                max_correlation = max(max_correlation, current_max_corr)

                # Если есть корреляция > 0.5 с любым существующим признаком, нельзя возвращать
                if current_max_corr > 0.5:
                    can_return = False

            # Если признак можно вернуть, добавляем в список
            if can_return:
                returnable_features.append({
                    'feature': feature,
                    'max_correlation': max_correlation
                })

    # Создаем DataFrame и сортируем по возрастанию максимальной корреляции
    if returnable_features:
        df_returnable = pd.DataFrame(returnable_features)
        df_returnable = df_returnable.sort_values('max_correlation', ascending=True)
        print("Признаки, которые можно вернуть:")
        print(df_returnable)
        print(f"\nВсего можно вернуть {len(df_returnable)} признаков")
    else:
        print("Нет признаков, которые можно вернуть (все имеют корреляцию > 0.5 с существующими признаками)")

```
